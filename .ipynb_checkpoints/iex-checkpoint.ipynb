```python
def baseURL(symbol):
    return "https://api.iextrading.com/1.0/stock/"+symbol+"/chart/date/"

baseURL("AAPL")
```

```python
from datetime import timedelta,datetime
import requests 
def getData(start,end,symbol):
    data = []
    curr = start
    while curr < end:
        print(curr.strftime('%Y%m%d'))
        res = requests.get(url =baseURL(symbol)+curr.strftime('%Y%m%d')+"?chartReset=true").json()
        print(len(res))
        data.extend(res)
        curr = curr + timedelta(days=1)
    return data
now = datetime.now()
data = getData(now-timedelta(days=30),now+timedelta(days=1),"AAPL")
```

# Creating the RNN model

```python
def trainModel(X,y):
    from livelossplot import PlotLossesKeras
    from keras import Sequential
    from keras.layers import LSTM, Activation, Dense, Dropout, Flatten
    model = Sequential()
    model.add(LSTM(1024,input_shape=(1,18),activation='relu',return_sequences=True))
    model.add(Dropout(0.2))
    model.add(Dense(512,activation='relu'))
    model.add(Dense(256,activation='relu'))
    model.add(Flatten())
    model.add(Dense(1))
    model.compile(loss="mse", optimizer="adam",metrics=['mae'])
    model.fit(X, y,epochs=50,validation_split=0.02,callbacks=[PlotLossesKeras()])
    return model
    
```

```python
model_low = trainModel(X,y_low)
```

```python
model_high = trainModel(X,y_high)
```
